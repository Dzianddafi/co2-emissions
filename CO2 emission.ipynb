# Introduction  

CO2 emissions from cars today continue to be a significant contributor to climate change. Most cars rely on internal combustion engines powered by fossil fuels, releasing carbon dioxide (CO2) into the atmosphere. These emissions are a primary driver of global warming, leading to more frequent extreme weather events and environmental disruptions. Efforts to reduce car CO2 emissions involve transitioning to electric and hybrid vehicles, improving fuel efficiency, and promoting sustainable transportation alternatives to mitigate this environmental impact.  
  

**Model_Year** = The establishment year of the car  
**Make** = Brand of the car  
**Model** = Car model  
**Vehicle_Class** = Type of the car  
**Engine_Size** = the volume of fuel and air that can be pushed through a car's cylinders and is measured in cubic centimetres (cc)  
**Cylinder**s = A cylinder is the power unit of an engine. itâ€™s the chamber where the gasoline is burned and turned into power.  
**Transmission** = acts as the medium that transmits power generated by the engine to the wheels via a mechanical system of gears and gear trains  
**Fuel_Consumption_in_City** = Amount of Litres used for each 100 km  
**Fuel_Consumption_in_City_Hwy** = Amount of Litres used in highways for each 100 km  
**Fuel_Consumption_comb** = Amount of litres used combinated for each 100 km  
**CO2_Emissions** = CO2 emitted by the car  
**smog_level** = This rating reflects vehicle tailpipe emissions that contribute to local and regional air pollution, creating problems such as smog, haze, and health issues.

# Correlogram  
  
A correlogram is a graphical representation or a table displaying correlation coefficients between variables in a dataset. Correlation coefficients measure the strength and direction of a linear relationship between two variables. The values range from -1 to 1, where -1 indicates a perfect negative linear relationship, 1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship.

```python
#Check the correlation of each features
sns.heatmap(data=df.corr(), annot=True)
```

# The linear regression model for CO2 emissions and fuel consumption

```python
# Linear regression model for CO2 emission 

X = np.array(df['Fuel_Consumption_in_City(L/100 km)']).reshape(-1,1)
y = np.array(df['CO2_Emissions'])

lr = LinearRegression()
lr.fit(X, y)
predictions = lr.predict(X)
plt.scatter(X, y, marker='*', c='red')
plt.plot(X, predictions, )
plt.title('Correlation between CO2 emissions and Fuel consumptions')
plt.xlabel('Fuel consumption (L/100 km)')
plt.ylabel('CO2 emissions')
plt.grid(True)
plt.show()
```

```python
# Calculating errors
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# r squared score
print(f'R squared score: {model.score(X_test, y_test)}')

# mean squared error
print(f'Root mean squared error score: {mean_squared_error(y_test, y_pred, squared=False)}')
```

[SUMMARY]  
  
The relation between CO2 emissions and fuel consumptions is parallel according to the graph and correlogram above. It has a very high number of correlation with **0.97** or **97%** indicating both variables have a nearly perfect positive correlation

# The linear regression model for CO2 emissions and smog level

```python
# Linear regression model for CO2 emission 
X = np.array(df['Smog_Level']).reshape(-1,1)
y = np.array(df['CO2_Emissions'])

lr = LinearRegression()
lr.fit(X, y)
predictions = lr.predict(X)
plt.scatter(X, y, marker='*', c='blue')
plt.plot(X, predictions)
plt.title('Correlation between CO2 emissions and smog level')
plt.xlabel('Smog level')
plt.ylabel('CO2 emissions')
plt.grid(True)
plt.show()
```

```python
# Calculating errors
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# r squared score
print(f'R squared score: {model.score(X_test, y_test)}')

# mean squared error
print(f'Root mean squared error score: {mean_squared_error(y_test, y_pred, squared=False)}')
```

[[SUMMARY]]  
  
The linear model showed above shows a negative correlation between CO2 emissions and smog level. This occurs because smog level refers to the density number of combination of air pollution and fog especially those pollution that produced by cars, industries, and other source of pollution. So, most likely the emmisions that produced by cars is just a small number comparing to other pollution source such as indistries.
